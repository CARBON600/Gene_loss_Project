```python
###now start gene prediction using prokka
def run_prokka(locustag,input,output):
    os.system(f"prokka --compliant --cpus 4 --outdir {output} --prefix {locustag} --locustag {locustag} {input}")
    
def prokka_annotation(pros,target_folder):
    wd = target_folder
    try:
        os.makedirs(f"{wd}/preflow/prokka_result", exist_ok=True)
    except OSError as err:
        if err.errno == os.errno.EEXIST and os.path.isdir(folder_path):
            pass
        else:
            raise
    params=[]
    pattern = os.path.join(f"{wd}/genome/", '*.fna')
    fna_files = glob.glob(pattern, recursive=True)
    for filepath in fna_files:
        dict_name=filepath.split("/")[-1].split(".")[0]
        params.append((dict_name,filepath,f"{wd}/preflow/prokka_result/{dict_name}"))
#prokka --compliant --cpus 4 --outdir /mnt/storage3/yfdai/material/ruegeria/GNM000011965_result --prefix GNM000011965 --locustag GNM000011965 /mnt/storage3/yfdai/material/ruegeria/GNM000011965.fna
    with multiprocessing.Pool(processes=pros) as pool:
        pool.starmap(run_prokka,params)
        pool.close
        pool.join
```

```python
### align single ortholog genes through mafft , globalpair separately 
def align_single_orthogroup(orthogroup_folder,align_folder):
    for filename in os.listdir(orthogroup_folder):
        filepath = os.path.join(orthogroup_folder, filename)
        if os.path.isfile(filepath):
            print(f"nohup /mnt/maple/ymchen/scripts/mafft/scripts/mafft --globalpair --maxiterate 1000 {filepath} > {align_folder}/{filename} &")
align_single_orthogroup("/mnt/maple/ymchen/gene_loss/preflow/ortho/Results_Sep11/Single_Copy_Orthologue_Sequences/","/mnt/maple/ymchen/gene_loss/preflow/msa")
```

```python
###revise msa result through trimal
def revise_msa_result(align_folder,revise_align_folder):
    for filename in os.listdir(align_folder):
        filepath = os.path.join(align_folder, filename)
        if os.path.isfile(filepath):
            os.system(f"trimal -in {filepath} -out {revise_align_folder}/{filename} -automated1")
revise_msa_result("/mnt/maple/ymchen/gene_loss/preflow/msa","/mnt/maple/ymchen/gene_loss/preflow/revised_msa")
```

```python
###merge msa result to prepare for the phylogenetic
def concatenate_sequences(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    sequence_dict = {}
    for filename in os.listdir(input_folder):
        filepath = os.path.join(input_folder, filename)

        if os.path.isfile(filepath):
            with open(filepath, 'r') as file:
                current_name = None
                for line in file:
                    line = line.strip()
                    if line.startswith('>'):
                        current_name = line[1:].rsplit("_",1)[0] 
                        if current_name not in sequence_dict:
                            sequence_dict[current_name] = []
                    else:
                        sequence_dict[current_name].append(line)
    
    for name, sequences in sequence_dict.items():
        output_path = os.path.join(output_folder, f"{name}.faa")
        with open(output_path, 'w') as output_file:
            output_file.write(f">{name}\n")
            output_file.write("".join(sequences) + "\n")

input_folder = "/mnt/maple/ymchen/gene_loss/preflow/revised_msa/"
output_folder = "/mnt/maple/ymchen/gene_loss/preflow/concated_msa_name"
concatenate_sequences(input_folder, output_folder)
```

```python
def get_all_file_paths(directory):
    file_paths = []
    for root, _, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            file_paths.append(file_path)
    return file_paths
def merge_fasta_files(input_files, output_file):
    with open(output_file, 'w') as outfile:
        for file in input_files:
            with open(file, 'r') as infile:
                outfile.write(infile.read())
                outfile.write('\n')  # 添加换行符以确保各个文件内容分开
input_files = get_all_file_paths("/mnt/maple/ymchen/gene_loss/preflow/concated_msa_name")
output_file = "/mnt/maple/ymchen/gene_loss/preflow/combine.fas"
merge_fasta_files(input_files, output_file)

print(f"文件已合并为 {output_file}")
```
